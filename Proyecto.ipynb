# Bienvenido a la parte de limpieza de datos donde se intentará ser lo mas breve posible en las explicaciones de los pasos a seguir

## primeramente se importan las librerias

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("movies_dataset.csv")
```

```python
df.head()
```

# se eliminan las columnas seleccionadas por el cliente

```python
df.drop(columns=["video", "imdb_id", "adult", "original_title", "vote_count", "poster_path", "homepage"], inplace=True) # se dropean las columnas 
```

# luego se verifica el tipo de dato que hay en las columnas que quedan

```python
df.info()
```

# se cambian los nulos por 0 #

```python
df['revenue'].fillna(value=0, inplace=True)
```

```python
df['budget'].fillna(value=0, inplace=True)
```

# se cambia el tipo de dato en la columna seleccionada ya que no estaban en tipo de fecha

```python
df['release_date'] = pd.to_datetime(df['release_date'], format='%Y-%m-%d', errors='coerce') # aca se obvian los errores por si hay valores que no sean tipo fecha
```

```python
df['release_date'].info()
```

# se verifican los nulos 

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['release_date'].isna().sum()
```

```python
df[df['release_date'].isna()] # se miran detalladamente los nullos
```

# se pasa a la eliminación de nulos y luego verificación

```python
df = df.dropna(subset=['release_date'])
```

```python
df['release_date'].isna().sum()
```

# se crea la columna de año y se verifica

```python
df['release_year'] = df['release_date'].dt.year
```

```python
df.head(5)
```

# se crea una columna con el valor dividido de "revenue" y "budget" usando "fillna" como filtro así no tira nulos, nombrada "return" con valores decimales

```python
df['budget'] = pd.to_numeric(df['budget'], downcast='float', errors='coerce')
```

```python
df['return'] = df['revenue'].divide(df['budget']).fillna(0)
```

```python
df['return'].head()
```

```python
df['return'].replace([np.inf, -np.inf], np.nan, inplace=True)  # se cambia el valor infinito por valor nullo
```

```python
df['return'].fillna(0, inplace=True) # se cambia el valor nullo por 0.00
```

```python
df['return'].head()
```

# Se desinada primeramente visualizando los datos para realizar las transformaciones

```python
df['belongs_to_collection'].head()
```

```python
df['belongs_to_collection'] = df['belongs_to_collection'].fillna('{}') 
# se rellenan los valores faltantes con diccionarios vacios
```

```python
df.head(5)
```

```python
df = df.reset_index(drop=True)
```

```python
df1= df['belongs_to_collection'].map(eval) # se genera df1 con la columna belongs_to_collection
```

```python
df1.head()
```

```python
df1 = df1.apply(pd.Series) # transforma los diccionarios en columnas/ tablas ( tener en cuenta que solo sirve en diccionarios)
```

```python
df1.head()
```

```python
df1.rename(columns={'id': 'id_collection', 'name': 'name_collection'}, inplace=True) # se renombran las columnas
df1.drop(columns=['poster_path', 'backdrop_path'], inplace=True) # se dropean las columnas que no tienen datos de importancia
```

```python
df1.head()
```

# en esta etapa como algunas columnas estan en tipo lista y con llaves de diccionario se realiza la siguiente transformación

```python
df2 =pd.json_normalize(df['genres'].map(eval)) 
```

```python
df2
```

```python
Id = []
Names= []
for i in range(8):
    nid = 'id'+ str(i)
    nam = 'name'+ str(i)
    Id.append(nid)
    Names.append(nam)
    df2[nid]= pd.json_normalize(df2[i])['id']
    df2[nam] = pd.json_normalize(df2[i])['name'] # como se visualiza arriba el la columna dentro del df2 contiene 8 columnas por eso el rango es el mismo
```

```python
df2.head()
```

```python
col= list(range(8)) # se crea una variable con el rango de la lista 
```

```python
df2.drop(columns=col, inplace=True) # para luego dropearlas ya que se desanidaron
```

```python
df2.head()
```

```python
df2['id_genres'] = df2[Id].apply(
    lambda x: ',' .join(x.dropna().astype(str)),
    axis = 1)
```

```python
df2['name_genres'] = df2[Names].apply(
    lambda x: ',' .join(x.dropna().astype(str)),
    axis = 1)
```

# una ves desanidadas las listas se crean 2 columnas en las cuales más adelante pasan desapercibidos los nulos

```python
df2['id_genres'].replace('', np.nan, inplace=True)
df2['name_genres'].replace('', np.nan, inplace=True)
```

```python
df2
```

# Quedando así la transformación

```python
df2.drop(columns=Id, inplace= True)
df2.drop(columns=Names, inplace= True)
df2
```

```python
df2.isnull().sum() # es opcion del usuario si desea o no eliminar los nullos
```
