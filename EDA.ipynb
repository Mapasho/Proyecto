## grafico para ver la cantidad de valores nullos

```python
plt.figure(figsize=(10, 8))
plt.imshow(df.isna(), aspect='auto',
           interpolation='nearest', cmap='gray')
plt.xlabel('Column Number')
plt.ylabel('Sampe Number')   
```

## grafico de barras para ver los valores faltantes con mas detalle

```python
# Filtrar las columnas con valores faltantes
mask = df.isna().any(axis=0)

# Calcular el porcentaje de valores faltantes y mostrar el gráfico
df.isna().mean()[mask].sort_values().plot(
    kind='bar', figsize=(15, 4),
    title='Porcentaje de valores faltantes por columna',
    ylabel='Relación de valores perdidos por columna'
)
```

```python
df.isnull().sum().sort_values(ascending=False)
```
